```python
import os
import cv2
import shutil
import imghdr
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.models import load_model
import numpy as np
import pandas as pd
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import matplotlib.pyplot as plt

# Step 1: Data Augmentation and Preprocessing
train_dir = './processed_split/train'
val_dir = './processed_split/val'

def preprocess_img(img):
    """Apply preprocessing to an image."""
    img = cv2.GaussianBlur(img, (5, 5), 0)
    img = tf.image.adjust_contrast(img, 2)
    return img

# Validate images before processing
def validate_images(directory):
    for root, _, files in os.walk(directory):
        for file in files:
            file_path = os.path.join(root, file)
            if not imghdr.what(file_path):
                print(f"Invalid image file detected and removed: {file_path}")
                os.remove(file_path)

validate_images(train_dir)
validate_images(val_dir)

# Data augmentation for training
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    horizontal_flip=True,
    vertical_flip=True,
    shear_range=0.2,
    fill_mode='nearest',
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    preprocessing_function=preprocess_img,
    brightness_range=[0.5, 1.5]
)

# Data normalization for validation
val_datagen = ImageDataGenerator(rescale=1./255)

# Load training and validation data
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary'
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(150, 150),
    batch_size=32,
    class_mode='binary'
)

# Debugging generator output
print("Sample data from train generator:")
sample_batch = next(train_generator)
print(f"Batch images shape: {sample_batch[0].shape}")
print(f"Batch labels: {sample_batch[1]}")

# Step 2: Building the CNN Model
def build_model():
    model = models.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
        layers.MaxPooling2D(2, 2),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D(2, 2),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D(2, 2),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D(2, 2),
        layers.Flatten(),
        layers.Dropout(0.5),
        layers.Dense(512, activation='relu'),
        layers.Dense(1, activation='sigmoid')
    ])

    model.compile(
        loss='binary_crossentropy',
        optimizer=tf.optimizers.Adam(),
        metrics=['accuracy']
    )

    return model

model = build_model()

# Step 3: Callbacks
callbacks_list = [
    callbacks.EarlyStopping(monitor='val_accuracy', patience=6, mode='max', verbose=1),
    callbacks.ModelCheckpoint(filepath='best_model.keras', monitor='val_accuracy', save_best_only=True, verbose=1)
]

# Train the model
history = model.fit(
    train_generator,
    steps_per_epoch=25,
    epochs=20,
    validation_data=val_generator,
    validation_steps=5,
    verbose=2,
    callbacks=callbacks_list
)

# Plot training history
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='lower right')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper right')
plt.show()

# Step 4: Evaluate the model
best_model = load_model('best_model.keras')
scores = best_model.evaluate(val_generator, verbose=1)
print(f"Validation Accuracy: {scores[1] * 100:.2f}%")
print(f"Validation Loss: {scores[0] * 100:.2f}%")

# Step 5: Predict on test data
def process_image(image_path):
    if not imghdr.what(image_path):
        print(f"Skipping invalid image file: {image_path}")
        return None
    img = load_img(image_path, target_size=(150, 150))
    img_array = img_to_array(img) / 255.0
    return np.expand_dims(img_array, axis=0)

test_dir = './test'
submission = []

for img_name in os.listdir(test_dir):
    img_path = os.path.join(test_dir, img_name)
    if img_name.lower().endswith(('jpg', 'jpeg', 'png')):
        img_array = process_image(img_path)
        if img_array is not None:
            prediction = best_model.predict(img_array)
            label = 1 if prediction[0][0] > 0.5 else 0
            submission.append({'id': img_name, 'label': label})

submission_df = pd.DataFrame(submission)
submission_df.to_csv('submission.csv', index=False)
print("Submission file created.")
```
